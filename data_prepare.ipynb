# 翔冠数据预处理

```python
import pandas as pd
import numpy as np
```

# 数据读取

## 图片数据

```python
picture_file = "data\picture.csv"
picture_data = pd.read_csv(picture_file)
picture_data = picture_data[picture_data['device_id'].isin(["xiangguanD3","xiangguanD4"])]
picture_data['picture_id'] = picture_data['id']
del picture_data['id']
del picture_data['update_time']
del picture_data['create_time']
df2 = pd.DataFrame((x.split(':') for x in picture_data['collect_time']),index=picture_data.index,columns=['hour','m','s'])
df3 = pd.merge(df2,picture_data,left_index=True,right_index=True)
df3.collect_time=df3.hour
del df3['hour'],df3['m'],df3['s']

picture_data = df3
del df3
picture_data




```

## 天气数据

```python
weather_file = "data\weather.csv"
weather_data = pd.read_csv(weather_file)
weather_data = weather_data[weather_data['device_id'].isin(["XGC00103","XGC00104"])]
weather_data = weather_data[~(weather_data['humidity']==0)] # 删掉没啥意义的行
weather_data['weather_id'] = weather_data['id']
del weather_data['id']
del weather_data['update_time']
del weather_data['wind_direction']
del weather_data['soil_salinity'] # 数据库中全是0
del weather_data['rainfall'] # 数据库中全是0
del weather_data['create_time'] # 数据库中全是0


weather_data = weather_data.replace(['XGC00104','XGC00103'],['xiangguanD4','xiangguanD3'])

df2 = pd.DataFrame((x.split(':') for x in weather_data['collect_time']),index=weather_data.index,columns=['hour','m','s'])
df3 = pd.merge(df2,weather_data,left_index=True,right_index=True)
df3.collect_time=df3.hour
del df3['hour'],df3['m'],df3['s']
weather_data = df3
del df3
weather_data
```

## 合并

```python
df = pd.merge(picture_data, weather_data,how='outer')
# df = df[~(df['weather_id'].isnull() | df['picture_id'].isnull())]
df = df.dropna() 
df
# df3.to_csv("test.csv")
```

## 添加标签

```python
from datetime import datetime
labels = [
    ["2021 05 1","2021 06 01",	"出苗"],
    ["2021 06 01","2021 06 21",	"苗期"],
    ["2021 06 21","2021 07 01",	"拔节期"],
    ["2021 07 01","2021 07 21",	"小喇叭口期"],
    ["2021 07 21","2021 07 31",	"抽丝期"],
    ["2021 07 31","2021 08 10",	"籽粒形成期"],
    ["2021 08 10","2021 08 20",	"乳熟期"],
    ["2021 08 20","2021 09 19",	"腊熟期"],
    ["2021 09 19","2021 10 19",	"完熟期"],
]
def get_label(input):
    # input : 14/5/2021 05
    input_date = datetime.strptime(input.split(' ')[0], "%d/%m/%Y")
    idx = 0
    for start,end,name in labels:
        start_date = datetime.strptime(start,"%Y %m %d")
        end_date = datetime.strptime(end,"%Y %m %d")
        # print(type(start_date) ,type(end_date) ,type(input_date) )
        if (input_date-start_date).days>=0 and (input_date-end_date).days<=0:
            return name,idx
        idx += 1
    return None,None



# get_label("14/5/2021 05")

df['label_name'] = df.apply(lambda x: get_label(x.collect_time)[0], axis=1)
df['label'] = df.apply(lambda x: get_label(x.collect_time)[1], axis=1)
```
